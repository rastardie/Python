```python
my_dict={'Dave': '001', 'Ava': '002', 'Joe': '003'} # method of creating a dictionalry (using curly brackets)
```

```python
my_dict['Dave']
```

```python
print(my_dict)
```

```python
print(my_dict.keys())   # print keys
```

```python
print(my_dict.values())    # print values
```

```python
print(my_dict.get('Ava'))
```

```python
for x in my_dict:    #retrives only the keys
    print(x)
```

```python
for x in my_dict.values():    #retrives only the values
    print(x)
```

```python
for x,y in my_dict.items():
    print(x,y)
```

```python
print(my_dict)
print('&&&&&&&&&&&&&&&&&&&&')
my_dict['Dave']='004'
my_dict['Chris']='003'
print(my_dict)
```

```python
print(my_dict)
my_dict.pop('Ava')
```

```python
my_dict.popitem() #removes the last item in the dictionary if no parameter is given 
```

```python
del my_dict['Dave']
```

```python
print(my_dict)
```
